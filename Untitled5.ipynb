```python
import pandas as pd
df1=pd.read_csv('train_dataset.csv')
df1.head()
```

```python
df1=df1[['Description','Browser_Used','Device_Used','Is_Response']]
df1.head()
```

```python
import pandas as pd
df2=pd.read_csv('test_dataset.csv')
df2.head()
```

```python
df2=df2[['Description','Browser_Used','Device_Used']]
df2.head()
```

```python
for i in range(0,30000):
    if(df1.iloc[i]['Is_Response']=='happy'):
        df1.iloc[i]['Is_Response']=1
    else:
        df1.iloc[i]['Is_Response']=0

df1.head(10)
```

```python
df1['Description']=df1.Description.map(lambda x:x.lower())
df1.head()
```

```python
df2['Description']=df2.Description.map(lambda x:x.lower())
df2.head()
```

```python
df1['Browser_Used']=df1.Browser_Used.map(lambda x:x.lower())
df1.head()
```

```python
df2['Browser_Used']=df2.Browser_Used.map(lambda x:x.lower())
df2.head()
```

```python
df1['Device_Used']=df1.Device_Used.map(lambda x:x.lower())
df1.head()
```

```python
df2['Device_Used']=df2.Device_Used.map(lambda x:x.lower())
df2.head()
```

```python
df1['Description'] = df1.Description.str.replace('[^\w\s]', '')  
df1.head()
#df.message.map(lambda x:print(x))
```

```python
df2['Description'] = df2.Description.str.replace('[^\w\s]', '')  
df2.head()
#df.message.map(lambda x:print(x))
```

```python
import nltk
```

```python
df1['Description'] = df1['Description'].apply(nltk.word_tokenize)  
df1.head()
```

```python
df2['Description'] = df2['Description'].apply(nltk.word_tokenize)  
df2.head()
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
df1['Description'] = df1['Description'].apply(lambda b: [stemmer.stem(a) for a in b])
df1.head()
```

```python
df2['Description'] = df2['Description'].apply(lambda x: [stemmer.stem(y) for y in x])
df2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
df1['Description'] = df1['Description'].apply(lambda b: ' '.join(b))

df1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
df2['Description'] = df2['Description'].apply(lambda x: ' '.join(x))

df2.head()
```

```python
count_vect = CountVectorizer()  
counts = count_vect.fit_transform(df1['Description'])  
```

```python
counts2 = count_vect.transform(df2['Description']) 
```

```python
print((counts))
```

```python
print((counts2))
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(counts,df1['Is_Response'],random_state=89)
```

```python
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train, y_train)  
```

```python
import numpy as np
predicted = model.predict(X_test)
```

```python
predicted1=model.predict(counts2)
```

```python
predicted1= pd.DataFrame(predicted1, columns=['Is_Response']).to_csv('submit.csv')
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predicted))  
```
